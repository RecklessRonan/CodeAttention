```python
def top_n_small_rank(n, score_dict):
    ''' returns keys which match the top n scores of values from a name:score dict'''
    lot = [(k,v) for k, v in score_dict.items()] #make list of tuple from scores dict
    nl = []
    while len(lot)> 0:
        nl.append(min(lot, key=lambda x: x[1]))
        lot.remove(nl[-1])
    return [i[0] for i in nl[0:n]]
```

```python
def value_smaller_than_n(n, score_dict):
    ''' returns keys which match the top n scores of values from a name:score dict'''
    lot = [(k,v) for k, v in score_dict.items()] #make list of tuple from scores dict
    nl = []
    while len(lot)> 0:
        minpair=min(lot, key=lambda x: x[1])
        if minpair[1]>=n:
            break
        nl.append(minpair)
        lot.remove(nl[-1])
    return [i[0] for i in nl[0:n]]
```
